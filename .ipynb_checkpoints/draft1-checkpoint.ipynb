## Exclude outliers:
For each column, first it computes the Z-score of each value in the column, relative to the column mean and standard deviation.
Then is takes the absolute of Z-score because the direction does not matter, only if it is below the threshold

```python
from scipy import stats
import numpy as np
clean_df = data[(np.abs(stats.zscore(data)) < 3).all(axis=1)]
ICD.display(clean_df)
```

## Correlation Matrix  
Generally a correlation coefficient is a statistical measure that reflect correlation between variables. 

```python
percent_chg = clean_df.pct_change()

corr = percent_chg.corr()
```

```python
# id correlation matrix 
# k : number of variables for heatmap 
k = 20
  
cols = corrmat.nlargest(k, 'id')['id'].index 
  
cm = np.corrcoef(data[cols].values.T) 
f, ax = plt.subplots(figsize =(12, 10)) 
  
sns.heatmap(cm, ax = ax, cmap ="YlGnBu", 
            linewidths = 0.1, yticklabels = cols.values,  
                              xticklabels = cols.values)
```
